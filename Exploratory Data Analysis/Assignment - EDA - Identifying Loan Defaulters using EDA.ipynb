##### 2.1 Delete Columns

```python
# from data profiling it's clear that few columns have null values.
# drop the columns with most nulls and the columns that are of no use.
```

```python
print(new.shape)
new.isnull().sum()*100/new.shape[0]
```

```python
# Delete columns if the missing values are quite significant in number.
# drop columns with more than 40% of its values as nulls.
temp=(new.isnull().sum()*100/new.shape[0]).reset_index().rename(columns={"index":"column_name",0:"missing %"})
drop_cols=list(temp[temp["missing %"]>=40]["column_name"])
new=new.drop(drop_cols,axis=1)
```

```python
print(new.shape)
new.isnull().sum()*100/new.shape[0]
```

```python
#drop cols EXT_SOURCE_2,EXT_SOURCE_3 as EXT_SOURCE_1 has already been dropped and no idea what this col exactly is
new.drop(columns= ['EXT_SOURCE_2','EXT_SOURCE_3'], inplace = True)
```

##### 2.2 Merge and Rename columns

```python
#Create a column for documents provided: count the documents provided by the client
new['docs_provided'] = new.loc[:,'FLAG_DOCUMENT_2':'FLAG_DOCUMENT_21'].sum(axis=1)
new.drop(columns = new.loc[:,'FLAG_DOCUMENT_2':'FLAG_DOCUMENT_21'].columns,inplace=True)
```

```python
#create a column to indicate wether phone number of any kind is provided or not by the client
#drop the extra columns of mobiles flags 
#rename the columns FLAG_CONT_MOBILE,FLAG_EMAIL to more meaningful
new['phone_provided'] = new.apply(lambda x: x['FLAG_MOBIL']+x['FLAG_EMP_PHONE']+x['FLAG_WORK_PHONE']+x['FLAG_PHONE'],axis=1)
new.drop(columns=['FLAG_MOBIL','FLAG_EMP_PHONE','FLAG_WORK_PHONE','FLAG_PHONE'],inplace=True)
new.rename(columns={'FLAG_CONT_MOBILE':'phone_working','FLAG_EMAIL':'email_provided'},inplace=True)
```

```python
# review the number of columns left after cleaning and handling the columns.
print(new.shape)
new.isnull().sum()*100/new.shape[0]
```

### 3. Fix the missing values in rows

```python
# Delete rows if the number of missing values is insignificant.
```

##### Drop rows with very less missing values

```python
# from the above step its clear that many columns have very less (<1%) of their values as nulls.
# Drop the rows in which these columns have nulls values.

temp=(new.isnull().sum()*100/new.shape[0]).reset_index().rename(columns={"index":"column_name",0:"missing %"})
list(temp[(temp["missing %"]>0) & (temp["missing %"]<10)]["column_name"])
new = new[~new.AMT_ANNUITY.isnull() ]
new = new[~new.AMT_GOODS_PRICE.isnull()]
new = new[~new.NAME_TYPE_SUITE.isnull()]
new = new[~new.CNT_FAM_MEMBERS.isnull()]
new = new[~new.OBS_30_CNT_SOCIAL_CIRCLE.isnull()]
new = new[~new.DEF_30_CNT_SOCIAL_CIRCLE.isnull()]
new = new[~new.OBS_60_CNT_SOCIAL_CIRCLE.isnull()]
new = new[~new.DEF_60_CNT_SOCIAL_CIRCLE.isnull()]
new = new[~new.DAYS_LAST_PHONE_CHANGE.isnull()]
new.shape
```

```python
# review the count of nulls in each column and see how many left with nulls at this point.
new.isnull().sum()
```

```python
# Columns with more than 40% were dropped and rows based on columns with <1% nulls were dropped.
# now impute the remaining rows or columns that are left with null values.
```

##### Handling the missing values in 'Occupation type'

```python
# 31% of values in this column are nulls. 
# Add mode to the missing values is not meaningful here, just indicate 'missing' instead
print("% of missing values in 'Occupation Type':",f'{100*new.OCCUPATION_TYPE.isnull().sum()/new.shape[0]}')
```

```python
# fill the missing values with 'missing'
new.OCCUPATION_TYPE.fillna('missing',inplace = True)
new.OCCUPATION_TYPE.isnull().sum()
```

##### Handling missing values in 'Credit Bureau cols'

```python
#fill with 0 or mode where null in below cols. as mean and medain might be continuous values which don't make sense here 
#AMT_REQ_CREDIT_BUREAU_HOUR, AMT_REQ_CREDIT_BUREAU_DAY, 
#AMT_REQ_CREDIT_BUREAU_WEEK, AMT_REQ_CREDIT_BUREAU_MON
#AMT_REQ_CREDIT_BUREAU_QRT, AMT_REQ_CREDIT_BUREAU_YEAR
new.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(new.AMT_REQ_CREDIT_BUREAU_HOUR.mode()[0],inplace = True)
new.AMT_REQ_CREDIT_BUREAU_DAY.fillna(new.AMT_REQ_CREDIT_BUREAU_DAY.mode()[0],inplace = True)
new.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(new.AMT_REQ_CREDIT_BUREAU_WEEK.mode()[0],inplace = True)
new.AMT_REQ_CREDIT_BUREAU_MON.fillna(new.AMT_REQ_CREDIT_BUREAU_MON.mode()[0],inplace = True)
new.AMT_REQ_CREDIT_BUREAU_QRT.fillna(new.AMT_REQ_CREDIT_BUREAU_QRT.mode()[0],inplace = True)
new.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(new.AMT_REQ_CREDIT_BUREAU_YEAR.mode()[0],inplace = True)
```
